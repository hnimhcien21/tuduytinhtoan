```python
#W5A3
from math import *
def isprime(n):
    if n < 2:
        return False
    for i in range(2, isqrt(n) + 1):
        if n % i == 0:
            return False
    return True

n = int(input())
if isprime(n):
    print('True')
else:
    print('False')
```

```python
# W5A4
def perfect_number(n):
    sum = 0
    for i in range(1, n):
        if n % i == 0:
            sum += i
    return sum == n

n = int(input())
if perfect_number(n):
    print('True')
else:
    print('False')
```

```python
# W5A5
def find(n):
    global k
    for i in range(len(n) + 1):
        if n[i] == k:
            return i + 1       
            break
    return -1

n = list(map(int, input().split()))
k = int(input())
print(find(n))

```

```python
# W5A6
def factorial(n):
    if n == 0 or n == 1:
        return 1
    for i in range(2, n + 1):
        n *= i
    return n

n = int(input())
print(factorial(n))
```

```python
#W5A7
def calculator(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return ('%.2f' % (a * b))
    elif op == '/':
        return ('%.2f' % (a / b))
 
a, op, b = input().split()
print(calculator(float(a), float(b), op))
```

```python
# W5A9
def sum(a):
    s = list(a)
    total = 0
    for i in range(len(s)):
        total += int(s[i])
    return total

n = input()
print(sum(n))
```

```python
# W5A10
def is_isomorphic(a, b):
    map1 = {}
    map2 = {}
    for x, y in zip(a, b):
        if (x in map1 and map1[x] != y) or (y in map2 and map2[y] != x):
            return False
        map1[x] = y
        map2[y] = x
    return True
a, b = input().split()
if is_isomorphic(a, b):
    print('true')
else:
    print('false')
```
