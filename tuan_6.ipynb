```python
#W6A4
s = input().split()
d = {}

for pair in s:
    key, value = pair.split(':')
    if key not in d:
        d[key] = [value]
    else:
        d[key].append(value)

print(d)


```

```python
#W6A5
a = list(map(int, input().split()))

d = {'positives': 0, 'negatives': 0, 'zeros': 0}

for x in a:
    if x > 0:
        d['positives'] += 1
    elif x < 0:
        d['negatives'] += 1
    else:
        d['zeros'] += 1

print(d)
```

```python
#W6A9
d1 = eval(input())
d2 = eval(input())

res = {}

for k in d1:
    if k in d2:
        res[k] = d1[k] + d2[k]
    else:
        res[k] = d1[k]

for k in d2:
    if k not in res:
        res[k] = d2[k]

res = dict(sorted(res.items()))

print(res)
```

```python
#W6A10
a = list(map(int, input().split()))
k = int(input())

res = []

n = len(a)
for i in range(n):
    for j in range(i + 1, n):
        if a[i] + a[j] == k:
            res.append((a[i], a[j]))

res.sort(key=lambda x: x[0])

print(res)
```

```python
#W6A12
a = list(map(int, input().split()))

cnt = {}
for x in a:
    cnt[x] = cnt.get(x, 0) + 1

max_freq = max(cnt.values())

res = min(x for x in cnt if cnt[x] == max_freq)

print(res)
```

```python
#W6A20
def tim_vi_tri(a, k):
    for i in range(len(a)):
        if a[i] == k:
            return i
    return -1
a = list(map(int,input().split()))
n = int(input())
print(tim_vi_tri(a, n))
```

```python
#W6A21
words = input().split()

pos = {}

for i in range(len(words)):
    w = words[i]
    if w not in pos:
        pos[w] = [i]
    else:
        pos[w].append(i)

# đổi list -> tuple
for w in pos:
    pos[w] = tuple(pos[w])

print(pos)
```
